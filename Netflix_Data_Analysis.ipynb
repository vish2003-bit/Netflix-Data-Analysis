```python
import pandas as pd

df = pd.read_csv('/netflix_titles.csv')

df.head()
```

```python
print("Dataset Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns.tolist())
```

```python
df['type'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar')

plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Count')

plt.show()
```

Insight 1:

Netflix contains significantly more Movies (6,131) than TV Shows (2,676).

This indicates that Netflix's content library is primarily focused on movies, which represent approximately 70% of the available titles.

```python
top_countries = df['country'].value_counts().head(10)

top_countries
```

```python
top_countries.plot(kind='bar')

plt.title('Top 10 Content Producing Countries')
plt.xlabel('Country')
plt.ylabel('Number of Titles')

plt.show()
```

## Insight 2

The United States is the leading content producer on Netflix, followed by India and the United Kingdom.

This highlights the strong influence of these countries in Netflix's global content library.

```python
import pandas as pd
import matplotlib.pyplot as plt

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['year_added'] = df['date_added'].dt.year

df['year_added'].value_counts().sort_index().plot(kind='line', marker='o')

plt.title('Netflix Content Added Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.grid(True)

plt.show()
```

Insight 3:

Netflix experienced significant growth in content additions between 2016 and 2020, indicating a period of rapid expansion of its content library.

## Conclusion

This analysis revealed that:

• Movies dominate Netflix's content library.
• The United States contributes the highest number of titles.
• Netflix rapidly expanded its catalog between 2016 and 2020.

These findings provide a basic understanding of Netflix's content distribution and growth trends.
